```python
import pafy
import time
import pytz
import datetime
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.common.exceptions import StaleElementReferenceException

from multiprocessing import Pool
```

```python
def get_video_urls(playlist_url):
    try:
        driver = webdriver.Chrome()
        driver.get(playlist_url)

        urls_done = set()
        last_updated = time.time()
        while time.time() - last_updated < 60:
            for elem in driver.find_elements_by_css_selector("#items a.ytd-thumbnail"):
                try:
                    href = elem.get_attribute('href')
                    if href not in urls_done:
                        yield href
                        urls_done.add(href)
                        last_updated = time.time()
                except StaleElementReferenceException as e:
                    continue
            driver.execute_script('var items = document.getElementById("items");'
                                  'while (items.firstChild) items.removeChild(items.firstChild);')
            driver.execute_script("window.scrollTo(0, 0); window.scrollTo(0, 10000000)")
            time.sleep(0.5)
    except Exception as e:
        print(e)
    finally:
        driver.close()
        driver.quit()
```

```python
with open('rtvnoord', 'w') as f: pass
url = 'https://www.youtube.com/user/rtvnoord/videos'
for i, x in enumerate(get_video_urls(url)):
    with open('rtvnoord', 'a') as f:
        f.write(x + '\n')
```

```python
import xml.etree.ElementTree as ET

def parse_xml(tree):
    words = []
    for p in next(x for x in tree.getchildren() if x.tag == 'body').getchildren():
        if p.tag == 'p':
            offset_time = int(p.get('t'))
            for s in p.getchildren():
                t = int(s.get('t', 0)) + offset_time
                if len(words) > 0:
                    words[-1]['to'] = t
                words.append({'from': t, 'word': s.text.strip(), 'ac': s.get('ac')})
            if len(p.getchildren()) == 0 and len(words) > 0:
                words[-1]['to'] = int(p.get('t'))
    return words

def get_captions(url, language_code='nl'):
    url = url.replace('\n', '')
    response = requests.get(url);
    data = response.text
    for x in data.split('\n'):
        if x.find('timed') >= 0:
            begin = x.find('ytplayer.config = ')
            end = x.find(';ytplayer.load')
            if begin >= 0 and end >=0:
                data = json.loads(x[x.find('ytplayer.config = ') + len('ytplayer.config = '):end])
                break
    else:
        return None
    captions = json.loads(data['args']['player_response'])['captions']['playerCaptionsTracklistRenderer']['captionTracks']

    for caption in captions:
        if caption['languageCode'] == language_code:
            caption_url = caption['baseUrl'] + '&fmt=srv3'
#             print(caption_url)
            caption_xml = requests.get(caption_url).text
#             print(caption_xml)
            tree = ET.fromstring(caption_xml)
            return parse_xml(tree)
```
